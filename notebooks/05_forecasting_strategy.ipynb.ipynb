### Forecasting Strategy

The forecasting strategy focuses on both aggregated business-level demand and detailed store-level demand.
A time-based trainâ€“test split is used to prevent data leakage.
Baseline forecasting models are established to ensure that advanced models provide meaningful performance improvements.


```python
import pandas as pd

df = pd.read_csv("../data/processed/rossmann_features.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
global_ts = (
    df.groupby('Date')['Sales']
      .sum()
      .reset_index()
      .sort_values('Date')
)
```

```python
STORE_ID = df.groupby('Store')['Date'].count().idxmax()

store_ts = (
    df[df['Store'] == STORE_ID]
    .sort_values('Date')
    [['Date', 'Sales']]
)
```

```python
split_date = '2015-06-01'

train_store = store_ts[store_ts['Date'] < split_date]
test_store  = store_ts[store_ts['Date'] >= split_date]
```

```python
train_store.to_csv("../data/processed/store_train.csv", index=False)
test_store.to_csv("../data/processed/store_test.csv", index=False)
```
