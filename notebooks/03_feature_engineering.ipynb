```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/rossmann_cleaned.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
df['Day'] = df['Date'].dt.day
df['IsWeekend'] = df['DayOfWeek'].isin([6, 7]).astype(int)
```

```python
df['StateHoliday'] = df['StateHoliday'].fillna('None')
df['IsStateHoliday'] = (df['StateHoliday'] != 'None').astype(int)
```

```python
df['SchoolHoliday'] = df['SchoolHoliday'].astype(int)
```

```python
df['Promo'] = df['Promo'].astype(int)
```

```python
df['IsPromo2Active'] = df['IsPromo2Active'].astype(int)
```

```python
assortment_map = {'a': 1, 'b': 2, 'c': 3}
df['AssortmentLevel'] = df['Assortment'].map(assortment_map)
```

```python
df['LogSales'] = np.log1p(df['Sales'])
```

```python
df = df.sort_values(['Store', 'Date'])
```

```python
df[['Sales', 'Promo', 'IsPromo2Active', 'IsStateHoliday']].describe()
```

```python
df.to_csv("../data/processed/rossmann_features.csv", index=False)
```
